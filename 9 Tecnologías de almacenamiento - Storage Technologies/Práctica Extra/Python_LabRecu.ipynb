```python
df["genres"] = df["genres"].str.replace("'", '"', regex=False)
```

```python
def texto_a_json(texto):
    try:
        objeto_json = json.loads(texto)
        return objeto_json
    except json.JSONDecodeError as e:
        print(f"Error al convertir el texto a JSON: {e}")
        return None

# Ejemplo de uso
texto = '[{"id": 1, "nombre": "Juan", "edad": 30}]'
objeto_json = texto_a_json(texto)
print(objeto_json)
print(objeto_json[0].get("nombre"))
```

```python
def select_random_element(json_array_str):
    try:
        if pd.isna(json_array_str):
            return "SenseGenere"
        json_array = json.loads(json_array_str)
        if json_array:
            elemento = random.choice(json_array)
            print(elemento)
            return elemento.get("name")
        else:
            return "SenseGenere"
    except (json.JSONDecodeError, TypeError, KeyError) as e:
        print(e)
        return "SenseGenere"

df['GenereAleatori'] = df['genres'].apply(select_random_element)
```
