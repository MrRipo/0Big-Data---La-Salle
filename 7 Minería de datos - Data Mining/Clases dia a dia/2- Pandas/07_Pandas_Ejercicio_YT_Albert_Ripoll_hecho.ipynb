8- Crea una nueva columna en el dataset con el continente de cada canal.

```python
df["Country"].value_counts()
```

```python
# Define a dictionary mapping countries to continents
country_to_continent = {
    'India': 'Asia',
    'United States': 'North America',
    'South Korea': 'Asia',
    'Brazil': 'South America',
    'Cyprus': 'Europe',
    'Sweden': 'Europe',
    'Ukraine': 'Europe',
    'Russia': 'Europe',
    'Canada': 'North America',
    'Argentina': 'South America',
    'United Kingdom': 'Europe',
    'Romania': 'Europe',
    'Chile': 'South America',
    'Mexico': 'North America',
    'El Salvador': 'North America',
    'United States (Puerto Rico)': 'North America'
}

# Clean spaces, dots, * and slashes that might be the 'Country' column in the DataFrame to match the dictionary
df['Country'] = df['Country'].str.replace(r'\[.*\]', '', regex=True).str.strip()

# Create a new 'Continent' column by mapping the values from the 'Country' column
df['Continent'] = df['Country'].map(country_to_continent)
df.head(3)
```
